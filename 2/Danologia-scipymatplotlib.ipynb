```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import chirp, sawtooth, square, gausspulse, welch, periodogram

t = np.linspace(0, 1, 1000)

chirp_signal = chirp(t, f0=10, f1=100, t1=1, method='linear')
sawtooth_signal = sawtooth(2 * np.pi * 5 * t)
square_signal = square(2 * np.pi * 5 * t)
gausspulse_signal = gausspulse(t, fc=5, bw=0.5)

chirp_signal_noise = chirp_signal + 0.5 * np.random.randn(1000)
sawtooth_signal_noise = sawtooth_signal + 0.5 * np.random.randn(1000)
square_signal_noise = square_signal + 0.5 * np.random.randn(1000)
gausspulse_signal_noise = gausspulse_signal + 0.5 * np.random.randn(1000)


plt.figure(figsize=(12, 8))

# Chirp Signal
plt.subplot(4, 4, 1)
plt.title('Chirp Signal')
plt.plot(t, chirp_signal)
plt.xlabel('Time')
plt.ylabel('Amplitude')

# Welch Spectrum
plt.subplot(4, 4, 2)
f, Pxx = welch(chirp_signal)
plt.plot(f, Pxx)
plt.title('Chirp Welch Spectrum')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Power')

# Periodogram
plt.subplot(4, 4, 3)
f, Pxx = periodogram(chirp_signal)
plt.plot(f, Pxx)
plt.title('Chirp Periodogram')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Power')

plt.subplot(4, 4, 4)
plt.plot(t, chirp_signal_noise)
plt.title('Chirp Noise')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.subplot(4, 4, 5)
plt.title('Sawtooth Signal')
plt.plot(t, sawtooth_signal)
plt.xlabel('Time')
plt.ylabel('Amplitude')


plt.subplot(4, 4, 6)
f, Pxx = welch(sawtooth_signal)
plt.plot(f, Pxx)
plt.title('Sawtooth Welch Spectrum')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Power')


plt.subplot(4, 4, 7)
f, Pxx = periodogram(sawtooth_signal)
plt.plot(f, Pxx)
plt.title('Sawtooth Periodogram')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Power')

plt.subplot(4, 4, 8)
plt.plot(t, sawtooth_signal_noise)
plt.title('Sawtooth Noise')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.subplot(4, 4, 9)
plt.title('Square Signal')
plt.plot(t, square_signal)
plt.xlabel('Time')
plt.ylabel('Amplitude')


plt.subplot(4, 4, 10)
f, Pxx = welch(square_signal)
plt.plot(f, Pxx)
plt.title('Square Welch Spectrum')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Power')


plt.subplot(4, 4, 11)
f, Pxx = periodogram(square_signal)
plt.plot(f, Pxx)
plt.title('Square Periodogram')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Power')

plt.subplot(4, 4, 12)
plt.plot(t, square_signal_noise)
plt.title('Square Noise')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.subplot(4, 4, 13)
plt.title('Gaussian Pulse Signal')
plt.plot(t, gausspulse_signal)
plt.xlabel('Time')
plt.ylabel('Amplitude')


plt.subplot(4, 4, 14)
f, Pxx = welch(gausspulse_signal)
plt.plot(f, Pxx)
plt.title('Gaussian Welch Spectrum')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Power')


plt.subplot(4, 4, 15)
f, Pxx = periodogram(gausspulse_signal)
plt.plot(f, Pxx)
plt.title('Gaussian Periodogram')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Power')

plt.subplot(4, 4, 16)
plt.plot(t, gausspulse_signal_noise)
plt.title('Gaussian Noise')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal
import math

t = np.linspace(0, 4 * math.pi, 100)
y = np.sin(t)
y2 = np.sin(2 * t) + np.sin(5 * t) + 2*np.cos(t)
y3 = y + np.random.randn(100) 
y4 = y2 + np.random.randn(100)

#plt.figure(figsize=(12, 12 ))
plt.subplot(4, 3, 1)
plt.title('y')
plt.plot(t, y)

plt.subplot(4, 3, 2)
plt.title('welch')
x, y = signal.welch(y)
plt.plot(x, y)

plt.subplot(4, 3, 3)
plt.title('periodogram')
x, y = signal.periodogram(y)
plt.plot(x, y)

plt.subplot(4, 3, 4)
plt.plot(t, y2)

plt.subplot(4, 3, 5)
x, y = signal.welch(y2)
plt.plot(x, y)

plt.subplot(4, 3, 6)
x, y = signal.periodogram(y2)
plt.plot(x, y)

plt.subplot(4, 3, 7)
plt.plot(t, y3)

plt.subplot(4, 3, 8)
x, y = signal.welch(y3)
plt.plot(x, y)

plt.subplot(4, 3, 9)
x, y = signal.periodogram(y3)
plt.plot(x, y)

plt.subplot(4, 3, 10)
plt.plot(t, y4)

plt.subplot(4, 3, 11)
x, y = signal.welch(y4)
plt.plot(x, y)

plt.subplot(4, 3, 12)
x, y = signal.periodogram(y4)
plt.plot(x, y)

plt.show()
```
